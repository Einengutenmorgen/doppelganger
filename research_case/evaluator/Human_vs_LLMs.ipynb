```python
# Visualizing Human vs LLMs' Evaluation scores on different similarity aspects

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the file contianing the evaluation score of Human and LLMs
df = pd.read_excel('File_Name.xlsx', sheet_name='Sheet_Name')

# Select the relevant columns
df_plot = df[['List_of_Relevant_columns']]

# Set the x-axis ticks
x = np.arange(len(df_plot['ID']))

# Set the width of the bars
width = 0.15

# Create the figure and axis
fig, ax = plt.subplots(figsize=(15, 6))

# Plot each column as a bar
ax.bar(x - 1.5 * width, df_plot['Human'], width, label='Human')
ax.bar(x - 0.5 * width, df_plot['Phi-3'], width, label='Phi-3')
ax.bar(x + 0.5 * width, df_plot['Llama3.2-1B'], width, label='Llama3.2-1B')
ax.bar(x + 1.5 * width, df_plot['Llama3.2-1B (CCI)'], width, label='Llama3.2-1B (CCI)')

# Set labels and title
ax.set_xlabel('ID')
ax.set_ylabel('Score')
ax.set_title('Comparison of Human and LLM Scores')
ax.set_xticks(x)
ax.set_xticklabels(df_plot['ID'])

# Add legend
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Absolute difference (Human Vs LLMs) calculation for the evalaution scores

df['Phi-3_Diff'] = abs(df['Human'] - df['Phi-3'])
df['Llama3.2-1B_Diff'] = abs(df['Human'] - df['Llama3.2-1B'])
df['Llama3.2-1B_CCI_Diff'] = abs(df['Human'] - df['Llama3.2-1B (CCI)'])

# Determine the minimum difference for each row
min_diff = df[['Phi-3_Diff', 'Llama3.2-1B_Diff', 'Llama3.2-1B_CCI_Diff']].min(axis=1)

# Identify all columns with the minimum difference
df['Closest_LLM'] = df.apply(lambda row: [col for col in ['Phi-3', 'Llama3.2-1B', 'Llama3.2-1B (CCI)']
                                         if row[col.replace(' (CCI)', '_CCI') + '_Diff'] == min_diff[row.name]], axis=1)

# Count how often each LLM is the closest
closest_counts = df['Closest_LLM'].explode().value_counts()

print(closest_counts)

# Plot a bar chart of the counts
plt.figure(figsize=(10,6))
plt.bar(closest_counts.index, closest_counts.values)
plt.xlabel('LLM Model')
plt.ylabel('Frequency as Closest')
plt.title('Frequency of Each LLM Being the Closest to Human Evaluation')
plt.show()
```
